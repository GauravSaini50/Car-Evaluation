```python
column_names = ['Buying', 'Maintainance', 'Doors','Persons','Lug_boot','Safety','Acceptance']


df = pd.read_csv('car.data', header=None, names=column_names)

print(df.head())
```

```python
df= df.drop(['Doors'],axis=1)
```

```python
df.Persons.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Seats in Cars")
plt.ylabel('Number of cars')
plt.xlabel('Number of Persons');
```

```python
df.Safety.value_counts().nlargest(40).plot(kind='bar', figsize=(12,5))
plt.title("Number of Cars are Safe")
plt.ylabel('Number of Cars')
plt.xlabel('Safety of Cars');
```

```python
df.Acceptance.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Cars can be Accepted")
plt.ylabel('Number of Cars')
plt.xlabel('Acceptance of Cars');
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Acceptance'], y=df['Safety'])
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['Acceptance'],y=df['Safety'])
```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Encode categorical columns
le = LabelEncoder()
df_encoded = df.apply(le.fit_transform)

# Plot heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Encoded Features")
plt.tight_layout()
plt.show()
```
